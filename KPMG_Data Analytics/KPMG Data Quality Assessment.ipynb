# EDA


```python
df3.head()
```

```python
df3['year'] = pd.DatetimeIndex(df3['DOB']).year
df3['month'] = pd.DatetimeIndex(df3['DOB']).month
```

```python
sns.distplot(df3['year'])
```

```python
df3[df3.year < 1950]
```

```python
df3['past_3_years_bike_related_purchases'] = df3['past_3_years_bike_related_purchases'].apply(pd.to_numeric)
```

```python
from matplotlib import rcParams

rcParams['figure.figsize'] = 11.7,8.27
```

```python
df_wealth=df3.groupby(['wealth_segment']).sum().reset_index()
df_wealth
```

```python
plt.figure(figsize=(18,8))
colors = sns.color_palette('flare')[2:6]
plt.pie(labels=df_wealth['wealth_segment'], x=df_wealth['past_3_years_bike_related_purchases'],pctdistance=0.85,explode=[0.01,0.01,0.01],colors=colors,autopct='%1.2f%%')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
plt.savefig('Wealth Segment vs Purchases.png',bbox_inches='tight')
```

```python
df_jobcat = df3.groupby(['job_industry_category']).sum().reset_index()
df_jobcat
```

```python
plt.figure(figsize=(18,8))
colors = sns.color_palette('viridis')
plt.pie(labels=df_jobcat['job_industry_category'], x=df_jobcat['past_3_years_bike_related_purchases'],pctdistance=0.85,colors=colors,autopct='%1.2f%%')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
plt.savefig('Job Industry vs Purchases.png',bbox_inches='tight')
```

```python
df_year = df3.groupby(['year']).sum().reset_index()
df_year.describe()
```

```python
bins= [1938,1958,1988,2002]
labels = ['Old','Middle','Young']
df_year['agegrp'] = pd.cut(df_year['year'], bins=bins, labels=labels, right=False)
df_year.head()
```

```python
sns.barplot('agegrp', 'past_3_years_bike_related_purchases', data=df_year,palette='magma')
plt.xlabel('Agegroup')
plt.ylabel('Past 3 Years Bike Related Purchases')
plt.savefig('Agegroup vs Purchases.png')
```

```python
df3["gender"]=df3["gender"].map({'Female':0,'Male':1,'Unknown':-1})
```

```python
df_gen = df3.groupby(['gender']).sum().reset_index()
df_gen
```

```python
sns.barplot('gender','past_3_years_bike_related_purchases', data=df_gen)
plt.savefig('Gender vs Purchases.png')
plt.xlabel('Gender')
plt.ylabel('Past 3 Years Bike Related Purchases')
```

```python
df_year2 = df3.groupby(['year','gender','past_3_years_bike_related_purchases']).size().reset_index()
df_year2.head()
```

```python
bins= [1938,1958,1988,2002]
labels = ['Old','Middle','Young']
df_year2['agegrp'] = pd.cut(df_year2['year'], bins=bins, labels=labels, right=False)
df_year2.head()
```

```python
sns.barplot('agegrp', 'past_3_years_bike_related_purchases',hue='gender',data=df_year2,palette='magma')
plt.xlabel('Agegroup')
plt.ylabel('Past 3 Years Bike Related Purchases')
plt.savefig('Genderwise Agegroup vs Purchases.png')
```

```python
df_state = df3.groupby(['state','gender','past_3_years_bike_related_purchases']).size().reset_index()
df_state.head()
```

```python
sns.barplot('state','past_3_years_bike_related_purchases',data=df_state,hue='gender')
plt.xlabel('State')
plt.ylabel('Past 3 Years Bike Related Purchases')
plt.savefig('State gendr division vs Purchases.png')
```
